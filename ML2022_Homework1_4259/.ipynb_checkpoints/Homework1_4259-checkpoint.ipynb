```python
#Machine Learning Homework_1 
#Author: Kosmas Apostolidis 4259

import pandas as pd
import numpy as np
from math import sqrt
from math import pi
from math import exp
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_fscore_support
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn import svm
from sklearn.naive_bayes import GaussianNB
```

```python
#Read the train csv file
all_colors = []
all_types  = []

train_df = pd.read_csv('train.csv')
test_df  = pd.read_csv('test.csv')
colors   = train_df['color'].iloc[:]
types    = train_df['type'].iloc[:]

for c in list(colors):
    if c not in all_colors:
        all_colors.append(c)
    
for t in list(types):
    if t not in all_types:
        all_types.append(t)
    
train_df = train_df.replace({"clear":1/6 , "green":2/6 , "black":3/6 , "white":4/6 , "blue":5/6 , "blood":6/6})
test_df  = test_df.replace({"clear":1/6 , "green":2/6 , "black":3/6 , "white":4/6 , "blue":5/6 , "blood":6/6})

print("All ghosts colors:"+str(all_colors)+"\n")
print("All ghosts types:"+str(all_types)+"\n")

print("Train_df\n"+str(train_df))
print("Test_df\n"+str(test_df))
```

```python
X_train = train_df[['bone_length','rotting_flesh','hair_length','has_soul','color']].iloc[:]
y_train = train_df['type'].iloc[:]
X_test  = test_df[['bone_length','rotting_flesh','hair_length','has_soul','color']].iloc[:]
precision_scores = []
recall_scores = []
f1_scores = []
```

```python
#Method 1 
#k-NN Nearest Neighbor using EuclideanDistance

number_of_neighbors = [1,3,5,10] #Number of neighbors that k-NN checks 

for n in number_of_neighbors:
    knn = KNeighborsClassifier(n_neighbors = n , metric = "euclidean")
    y_train_pred = knn.fit(X_train,y_train).predict(X_train)
    y_test_pred  = knn.fit(X_train,y_train).predict(X_test) #Predict the type of the monster
    pr , r , f , s = precision_recall_fscore_support(y_train,y_train_pred,average = 'weighted')
    precision_scores.append(pr)
    recall_scores.append(r)
    f1_scores.append(f)
    
knn_metrics_df = pd.DataFrame({"number_of_neighbors":number_of_neighbors,"accuracy":[0.68809,0.69943,0.70510,0.71266],"precision":precision_scores,"recall":recall_scores,"f1":f1_scores}).set_index('number_of_neighbors')
print(knn_metrics_df)
knn_metrics_df = knn_metrics_df.reset_index()
fig,ax = plt.subplots(1,4,figsize=(20,5))
knn_metrics_df.plot(x = 'number_of_neighbors' , y = 'accuracy',ax=ax[0])
knn_metrics_df.plot(x = 'number_of_neighbors' , y = 'precision',ax=ax[1])
knn_metrics_df.plot(x = 'number_of_neighbors' , y = 'recall' , ax = ax[2])
knn_metrics_df.plot(x = 'number_of_neighbors' , y = 'f1' , ax = ax[3])
```

```python
#Method 3
#Support Vector Machines using linear and rbf(Gaussian) kernels(SVM)

linear_svc_accuracy_score = 0.73724 #Acuracy score given by kaggle for my linear_svc classification
linear_svc = svm.SVC(kernel = 'linear' , decision_function_shape = 'ovr')
y_train_pred = linear_svc.fit(X_train , y_train).predict(X_train)
y_test_pred  = linear_svc.fit(X_train , y_train).predict(X_test)
pr , r , f , s = precision_recall_fscore_support(y_train,y_train_pred,average = 'weighted')


#Into comment this line because every time you run this line a csv is created.It is used only for uploading into kaggle site
#linear_svc_df = pd.DataFrame({"id":ids,"type":y_test_pred}).to_csv('linear_svc_predictions.csv', index = False)

linear_svc_metrics_df = pd.DataFrame({"accuracy":[linear_svc_accuracy_score],"precision":[pr],"recall":[r],"f1":[f]})
print("Support Vector Machine with linear kernel with one-versus-all decision_function_shape")
print(linear_svc_metrics_df)


precision_scores.clear()
recall_scores.clear()
f1_scores.clear()
gamma = [0.0001,0.001,0.01,0.08,0.09]

for g in gamma:
    rbf_svc = svm.SVC(kernel = 'rbf' , gamma = g , decision_function_shape = 'ovr')
    y_train_pred = rbf_svc.fit(X_train , y_train).predict(X_train)
    y_test_pred  = rbf_svc.fit(X_train , y_train).predict(X_test)
    pr , r , f , s = precision_recall_fscore_support(y_train,y_train_pred,average = 'weighted',labels=np.unique(y_train_pred))
    precision_scores.append(pr)
    recall_scores.append(r)
    f1_scores.append(f)


rbf_svc_metrics_df = pd.DataFrame({"gamma":gamma,"accuracy":[0.32325,0.32325,0.32325,0.72589,0.74102],"precision":precision_scores,"recall":recall_scores,"f1":f1_scores}).set_index('gamma')
print("\n\n")
print("Support Vector Machine with Radial Basis Function (RBF) kernel with one-versus-all decision_function_shape")
print(rbf_svc_metrics_df)
rbf_svc_metrics_df = rbf_svc_metrics_df.reset_index()
fig,ax = plt.subplots(1,4,figsize=(20,5))
rbf_svc_metrics_df.plot(x = 'gamma' , y = 'accuracy',ax=ax[0])
rbf_svc_metrics_df.plot(x = 'gamma' , y = 'precision',ax=ax[1])
rbf_svc_metrics_df.plot(x = 'gamma' , y = 'recall' , ax = ax[2])
rbf_svc_metrics_df.plot(x = 'gamma' , y = 'f1' , ax = ax[3])
```

```python
#Method 4
#Implementation of Naive Bayes Classifier,but hardcoded without the usage of any helping library
monster_types = train_df['type'].iloc[:]

total_number_of_records  = len(train_df) 
 
sum_ghoul_bone_length    = float(train_df[train_df['type'] == "Ghoul"].groupby('type')['bone_length'].sum()) #Group train_df by type if type == 'Ghoul' and calculate the total sum of bone_length 
sum_ghoul_rotting_flesh  = float(train_df[train_df['type'] == "Ghoul"].groupby('type')['rotting_flesh'].sum())#Group train_df by type if type == 'Ghoul' and calculate the total sum of rotting_flesh 
sum_ghoul_hair_length    = float(train_df[train_df['type'] == "Ghoul"].groupby('type')['hair_length'].sum())#Group train_df by type if type == 'Ghoul' and calculate the total sum of hair_length 
sum_ghoul_has_soul       = float(train_df[train_df['type'] == "Ghoul"].groupby('type')['has_soul'].sum())#Group train_df by type if type == 'Ghoul' and calculate the total sum of has_soul
sum_ghoul_color          = float(train_df[train_df['type'] == "Ghoul"].groupby('type')['color'].sum())#Group train_df by type if type == 'Ghoul' and calculate the total sum of color 

sum_goblin_bone_length   = float(train_df[train_df['type'] == "Goblin"].groupby('type')['bone_length'].sum())#Group train_df by type if type == 'Goblin' and calculate the total sum of bone_length 
sum_goblin_rotting_flesh = float(train_df[train_df['type'] == "Goblin"].groupby('type')['rotting_flesh'].sum())#Group train_df by type if type == 'Goblin' and calculate the total sum of rotting_flesh 
sum_goblin_hair_length   = float(train_df[train_df['type'] == "Goblin"].groupby('type')['hair_length'].sum())#Group train_df by type if type == 'Goblin' and calculate the total sum of hair_length  
sum_goblin_has_soul      = float(train_df[train_df['type'] == "Goblin"].groupby('type')['has_soul'].sum())#Group train_df by type if type == 'Goblin' and calculate the total sum of has_soul  
sum_goblin_color         = float(train_df[train_df['type'] == "Goblin"].groupby('type')['color'].sum())#Group train_df by type if type == 'Goblin' and calculate the total sum of color  

sum_ghost_bone_length    = float(train_df[train_df['type'] == "Ghost"].groupby('type')['bone_length'].sum())#Group train_df by type if type == 'Ghost' and calculate the total sum of bone_length 
sum_ghost_rotting_flesh  = float(train_df[train_df['type'] == "Ghost"].groupby('type')['rotting_flesh'].sum())#Group train_df by type if type == 'Ghost' and calculate the total sum of rotting_flesh 
sum_ghost_hair_length    = float(train_df[train_df['type'] == "Ghost"].groupby('type')['hair_length'].sum())#Group train_df by type if type == 'Ghost' and calculate the total sum of hair_length 
sum_ghost_has_soul       = float(train_df[train_df['type'] == "Ghost"].groupby('type')['has_soul'].sum())#Group train_df by type if type == 'Ghost' and calculate the total sum of has_soul 
sum_ghost_color          = float(train_df[train_df['type'] == "Ghost"].groupby('type')['color'].sum())#Group train_df by type if type == 'Ghost' and calculate the total sum of color 

number_of_records_with_class_ghoul  = sum([1 for type_ in monster_types if type_ == "Ghoul"])#Total number of records with class Ghoul
number_of_records_with_class_goblin = sum([1 for type_ in monster_types if type_ == "Goblin"])#Total number of records with class Goblin
number_of_records_with_class_ghost  = sum([1 for type_ in monster_types if type_ == "Ghost"])#Total number of records with class Ghost

pr_ghoul  = number_of_records_with_class_ghoul  / total_number_of_records #The probability the type of a specific monster to be ghoul
pr_goblin = number_of_records_with_class_goblin / total_number_of_records #The probability the type of a specific monster to be goblin
pr_ghost  = number_of_records_with_class_ghost  / total_number_of_records #The probability the type of a specific monster to be ghost

ghoul_mean_bl  = sum_ghoul_bone_length   / number_of_records_with_class_ghoul #Ghoul mean bone_length value 
ghoul_mean_rf  = sum_ghoul_rotting_flesh / number_of_records_with_class_ghoul #Ghoul mean rotting_flesh value
ghoul_mean_hl  = sum_ghoul_hair_length   / number_of_records_with_class_ghoul #Ghoul mean hair_length value
ghoul_mean_hs  = sum_ghoul_has_soul      / number_of_records_with_class_ghoul #Ghoul mean has_sooul value
ghoul_mean_c   = sum_ghoul_color         / number_of_records_with_class_ghoul #Ghoul mean color value

#Mean values for all attributes of Goblin type
goblin_mean_bl  = sum_goblin_bone_length   / number_of_records_with_class_goblin #goblin mean bone_length value 
goblin_mean_rf  = sum_goblin_rotting_flesh / number_of_records_with_class_goblin #goblin mean rotting_flesh value
goblin_mean_hl  = sum_goblin_hair_length   / number_of_records_with_class_goblin #goblin mean hair_length value
goblin_mean_hs  = sum_goblin_has_soul      / number_of_records_with_class_goblin #goblin mean has_sooul value
goblin_mean_c   = sum_goblin_color         / number_of_records_with_class_goblin #goblin mean color valuegoblin

#Mean values for all attributes of Ghost type
ghost_mean_bl  = sum_ghost_bone_length   / number_of_records_with_class_ghost #ghost mean bone_length value 
ghost_mean_rf  = sum_ghost_rotting_flesh / number_of_records_with_class_ghost #ghost mean rotting_flesh value
ghost_mean_hl  = sum_ghost_hair_length   / number_of_records_with_class_ghost #ghost mean hair_length value
ghost_mean_hs  = sum_ghost_has_soul      / number_of_records_with_class_ghost #ghost mean has_sooul value
ghost_mean_c   = sum_ghost_color         / number_of_records_with_class_ghost #ghost mean color value 

ghoul_sum_of_bl_minus_mean_bl_square  = sum([(bl - ghoul_mean_bl)**2 for bl in train_df[train_df['type'] == "Ghoul"]['bone_length']])       #Σ(ghoul_bone_length[i]−μ)^2
ghoul_sum_of_rf_minus_mean_rf_square  = sum([(rf - ghoul_mean_rf)**2 for rf in train_df[train_df['type'] == "Ghoul"]['rotting_flesh']])     #Σ(ghoul_rotting_flesh[i]−μ)^2
ghoul_sum_of_hl_minus_mean_hl_square  = sum([(hl - ghoul_mean_hl)**2 for hl in train_df[train_df['type'] == "Ghoul"]['hair_length']])       #Σ(ghoul_hair_length[i]−μ)^2
ghoul_sum_of_hs_minus_hs_square       = sum([(hs - ghoul_mean_hs)**2 for hs in train_df[train_df['type'] == "Ghoul"]['has_soul']])          #Σ(ghoul_has_soul[i]−μ)^2
ghoul_sum_of_c_minus_mean_c_square    = sum([(c  - ghoul_mean_c) **2 for c  in train_df[train_df['type'] == "Ghoul"]['color']])             #Σ(ghoul_color[i]−μ)^2

goblin_sum_of_bl_minus_mean_bl_square = sum([(bl - goblin_mean_bl)**2 for bl in train_df[train_df['type'] == "Goblin"]['bone_length']])     #Σ(goblin_bone_length[i]−μ)^2
goblin_sum_of_rf_minus_mean_rf_square = sum([(rf - goblin_mean_rf)**2 for rf in train_df[train_df['type'] == "Goblin"]['rotting_flesh']])   #Σ(goblin_rotting_flesh[i]−μ)^2
goblin_sum_of_hl_minus_mean_hl_square = sum([(hl - goblin_mean_hl)**2 for hl in train_df[train_df['type'] == "Goblin"]['hair_length']])     #Σ(goblin_hair_length[i]−μ)^2
goblin_sum_of_hs_minus_hs_square      = sum([(hs - goblin_mean_hs)**2 for hs in train_df[train_df['type'] == "Goblin"]['has_soul']])        #Σ(goblin_has_soul[i]−μ)^2
goblin_sum_of_c_minus_mean_c_square   = sum([(c  - goblin_mean_c)**2  for c  in train_df[train_df['type'] == "Goblin"]['color']])           #Σ(goblin_color[i]−μ)^2

ghost_sum_of_bl_minus_mean_bl_square  = sum([(bl - ghost_mean_bl)**2 for bl in train_df[train_df['type'] == "Ghost"]['bone_length']])       #Σ(ghost_bone_length[i]−μ)^2
ghost_sum_of_rf_minus_mean_rf_square  = sum([(rf - ghost_mean_rf)**2 for rf in train_df[train_df['type'] == "Ghost"]['rotting_flesh']])     #Σ(ghost_rotting_flesh[i]−μ)^2
ghost_sum_of_hl_minus_mean_hl_square  = sum([(hl - ghost_mean_hl)**2 for hl in train_df[train_df['type'] == "Ghost"]['hair_length']])       #Σ(ghost_hair_length[i]−μ)^2
ghost_sum_of_hs_minus_hs_square       = sum([(hs - ghost_mean_hs)**2 for hs in train_df[train_df['type'] == "Ghost"]['has_soul']])          #Σ(ghost_has_soul[i]−μ)^2
ghost_sum_of_c_minus_mean_c_square    = sum([(c  - ghost_mean_c)**2  for c  in train_df[train_df['type'] == "Ghost"]['color']])             #Σ(ghost_color[i]−μ)^2

#Variance values for all attributes of Ghoul type
ghoul_var_bl  = ghoul_sum_of_bl_minus_mean_bl_square / (number_of_records_with_class_ghoul - 1)
ghoul_var_rf  = ghoul_sum_of_rf_minus_mean_rf_square / (number_of_records_with_class_ghoul - 1)
ghoul_var_hl  = ghoul_sum_of_hl_minus_mean_hl_square / (number_of_records_with_class_ghoul - 1)
ghoul_var_hs  = ghoul_sum_of_hs_minus_hs_square      / (number_of_records_with_class_ghoul - 1)
ghoul_var_c   = ghoul_sum_of_c_minus_mean_c_square   / (number_of_records_with_class_ghoul - 1)

#Variance values for all attributes of Goblin type
goblin_var_bl = goblin_sum_of_bl_minus_mean_bl_square / (number_of_records_with_class_goblin - 1)
goblin_var_rf = goblin_sum_of_rf_minus_mean_rf_square / (number_of_records_with_class_goblin - 1)
goblin_var_hl = goblin_sum_of_hl_minus_mean_hl_square / (number_of_records_with_class_goblin - 1)
goblin_var_hs = goblin_sum_of_hs_minus_hs_square      / (number_of_records_with_class_goblin - 1)
goblin_var_c  = goblin_sum_of_c_minus_mean_c_square   / (number_of_records_with_class_goblin - 1)

#Variance values for all attributes of Ghost type
ghost_var_bl  = ghost_sum_of_bl_minus_mean_bl_square / (number_of_records_with_class_ghost - 1)
ghost_var_rf  = ghost_sum_of_rf_minus_mean_rf_square / (number_of_records_with_class_ghost - 1)
ghost_var_hl  = ghost_sum_of_hl_minus_mean_hl_square / (number_of_records_with_class_ghost - 1)
ghost_var_hs  = ghost_sum_of_hs_minus_hs_square      / (number_of_records_with_class_ghost - 1)
ghost_var_c   = ghost_sum_of_c_minus_mean_c_square   / (number_of_records_with_class_ghost - 1)

print("Total number of records in test dataset: "+str(total_number_of_records))
print("Number of records with class Ghoul: "+str(number_of_records_with_class_ghoul))
print("Number of records with class Goblin: "+str(number_of_records_with_class_goblin))
print("Number of records with class Ghost: "+str(number_of_records_with_class_ghost))
print("P(ghoul): "+str(pr_ghoul))
print("P(goblin): "+str(pr_goblin))
print("P(ghost): "+str(pr_ghost))

ghoul_metrics_df = pd.DataFrame({"bone_length"  :[ghoul_mean_bl,ghoul_var_bl] , 
                                 "rotting_flesh":[ghoul_mean_rf,ghoul_var_rf] , 
                                 "hair_length"  :[ghoul_mean_hl,ghoul_var_hl] , 
                                 "has_soul"     :[ghoul_mean_hs,ghoul_var_hs] , 
                                 "color"        :[ghoul_mean_c,ghoul_var_c]}).rename(index = {0:"mean" , 1:"variance"})

goblin_metrics_df = pd.DataFrame({"bone_length"  :[goblin_mean_bl,goblin_var_bl] ,
                                  "rotting_flesh":[goblin_mean_rf,goblin_var_rf] , 
                                  "hair_length"  :[goblin_mean_hl,goblin_var_hl] , 
                                  "has_soul"     :[goblin_mean_hs,goblin_var_hs] , 
                                  "color"        :[goblin_mean_c,goblin_var_c]}).rename(index = {0:"mean" , 1:"variance"})

ghost_metrics_df = pd.DataFrame({"bone_length"  :[ghost_mean_bl,ghost_var_bl] , 
                                 "rotting_flesh":[ghost_mean_rf,ghost_var_rf] , 
                                 "hair_length"  :[ghost_mean_hl,ghost_var_hl] , 
                                 "has_soul"     :[ghost_mean_hs,ghost_var_hs] , 
                                 "color"        :[ghost_mean_c,ghost_var_c]}).rename(index = {0:"mean" , 1:"variance"})
print("\nGhoul metrics")
print(ghoul_metrics_df)
print("\n\nGoblin metrics")
print(goblin_metrics_df)
print("\n\nGhost metrics")
print(ghost_metrics_df)
```

```python
#Here the conditional probabilities are calculated for given test record
#The highest probability is the class/type of the test record

y_test_pred = []

#The probability density function for the Gaussian/Normal distribution
def normal_distribution_PDF(x,μ,σ_square):
    return (1/(sqrt(2 * pi * σ_square))) * exp((- (x - μ)**2) / (2 * σ_square))
    
#Calculate the conditional probability of the test record given a type
def calculate_conditional_probability(type_,bl,rf,hl,hs,c):
    if type_  == "Ghoul":
        return pr_ghoul*normal_distribution_PDF(bl,ghoul_mean_bl,ghoul_var_bl)*normal_distribution_PDF(rf,ghoul_mean_rf,ghoul_var_rf)*normal_distribution_PDF(hl,ghoul_mean_hl,ghoul_var_hl)*normal_distribution_PDF(hs,ghoul_mean_hs,ghoul_var_hs)*normal_distribution_PDF(c,ghoul_mean_c,ghoul_var_c)
    elif type_ == "Goblin":
        return pr_goblin*normal_distribution_PDF(bl,goblin_mean_bl,goblin_var_bl)*normal_distribution_PDF(rf,goblin_mean_rf,goblin_var_rf)*normal_distribution_PDF(hl,goblin_mean_hl,goblin_var_hl)*normal_distribution_PDF(hs,goblin_mean_hs,goblin_var_hs)*normal_distribution_PDF(c,goblin_mean_c,goblin_var_c)
    else:
        return pr_ghost*normal_distribution_PDF(bl,ghost_mean_bl,ghost_var_bl)*normal_distribution_PDF(rf,ghost_mean_rf,ghost_var_rf)*normal_distribution_PDF(hl,ghost_mean_hl,ghost_var_hl)*normal_distribution_PDF(hs,ghost_mean_hs,ghost_var_hs)*normal_distribution_PDF(c,ghost_mean_c,ghost_var_c)

#for every attribute of every test record calculate the pdf for every type
for i in range(len(test_df)):
    bone_length   = test_df['bone_length'].iloc[i]
    rotting_flesh = test_df['rotting_flesh'].iloc[i]
    hair_length   = test_df['hair_length'].iloc[i]
    has_soul      = test_df['has_soul'].iloc[i]
    color         = test_df['color'].iloc[i]
    
    p_type_is_ghoul_for_given_bone_length    = normal_distribution_PDF(bone_length,ghoul_mean_bl,ghoul_var_bl)
    p_type_is_ghoul_for_given_rotting_flesh  = normal_distribution_PDF(rotting_flesh,ghoul_mean_rf,ghoul_var_rf)
    p_type_is_ghoul_for_given_hair_length    = normal_distribution_PDF(hair_length,ghoul_mean_hl,ghoul_var_hl)
    p_type_is_ghoul_for_given_has_soul       = normal_distribution_PDF(has_soul,ghoul_mean_hs,ghoul_var_hs)
    p_type_is_ghoul_for_given_color          = normal_distribution_PDF(color,ghoul_mean_c,ghoul_var_c)
    
    p_type_is_goblin_for_given_bone_length   = normal_distribution_PDF(bone_length,goblin_mean_bl,goblin_var_bl)
    p_type_is_goblin_for_given_rotting_flesh = normal_distribution_PDF(rotting_flesh,goblin_mean_rf,goblin_var_rf)
    p_type_is_goblin_for_given_hair_length   = normal_distribution_PDF(hair_length,goblin_mean_hl,goblin_var_hl)
    p_type_is_goblin_for_given_has_soul      = normal_distribution_PDF(has_soul,goblin_mean_hs,goblin_var_hs)
    p_type_is_goblin_for_given_color         = normal_distribution_PDF(color,goblin_mean_c,goblin_var_c)
    
    p_type_is_ghost_for_given_bone_length    = normal_distribution_PDF(bone_length,ghost_mean_bl,ghost_var_bl) 
    p_type_is_ghost_for_given_rotting_flesh  = normal_distribution_PDF(rotting_flesh,ghost_mean_rf,ghost_var_rf)
    p_type_is_ghost_for_given_hair_length    = normal_distribution_PDF(hair_length,ghost_mean_hl,ghost_var_hl)
    p_type_is_ghost_for_given_has_soul       = normal_distribution_PDF(has_soul,ghost_mean_hs,ghost_var_hs)
    p_type_is_ghost_for_given_color          = normal_distribution_PDF(color,ghost_mean_c,ghost_var_c)
    
    p_type_is_ghoul_given_attributes  = calculate_conditional_probability("Ghoul",bone_length,rotting_flesh,hair_length,has_soul,color)
    p_type_is_goblin_given_attributes = calculate_conditional_probability("Goblin",bone_length,rotting_flesh,hair_length,has_soul,color)
    p_type_is_ghost_given_attributes  = calculate_conditional_probability("Ghost",bone_length,rotting_flesh,hair_length,has_soul,color)
    
    if p_type_is_ghoul_given_attributes > p_type_is_goblin_given_attributes and p_type_is_ghoul_given_attributes > p_type_is_ghost_given_attributes:
        y_test_pred.append("Ghoul")
    elif p_type_is_goblin_given_attributes > p_type_is_ghoul_given_attributes and p_type_is_goblin_given_attributes > p_type_is_ghost_given_attributes:
        y_test_pred.append("Goblin")
    else:
        y_test_pred.append("Ghost")
```
